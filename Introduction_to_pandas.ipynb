Excluir linhas e colunas

```python
#para excluir a coluna quq quero, passo o venda df + o eixo (1 é da coluna e 0 da linha)
vendas_df = vendas_df.drop("Imposto", axis=1)
display(vendas_df)
```

```python
#para excluir a linha zero
#se não passar a informação do eixo, automaticamente o pandas vai deletar a linha
vendas_df = vendas_df.drop(0, axis=0)
display(vendas_df)
```

COMANDOS IMPORTANTE PARA TRATAMENTO DE DADOS
Valores Vazios
- Deletar linhas/colunas vazias
- Deletar linhas que possuem valores vazios
- Preencher valores vazios (média e último valor)

```python
#deletar linhas e colunas completamente vazias
#o dropna exclui valores vazios
#how = all, se passar somente esse parâmetro ele vai excluir linhas e colunas que forem completamentes vazios
#e se quiser excluir as colunas vazias, precisa incluir o paramêtro colunas = axis = 1

#vendas_df = vendas_df.dropna(how='all', axis=1)
```

```python
#deletar linhas que possuem pelo menos 1 valor vazio

#vendas_df = vendas_df.dropna()
```

```python
# preencher valores vazios e existem duas formas:
# preencher com a média da coluna
# vendas_df['Comissão'] -- novo valor 
# vai ser o antigo valor + fillna
# fillna(0) iria preencher todos os valores vazios com zero, ou qualquer outro número que for colocado entre parêntese
# fillna(vendas_df['Comissão'].mean) -- signfica que ele vai preencher todos os valores vazios com a média dis valores da comissão

vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean())
display(vendas_df)
```

```python
# preencher com o último valor
#ele vai preencer com o valor acima dele
vendas_df = vendas_df.ffill()
display(vendas_df)
```

Calcular Indicadores
- Groupby
- Value Counts

```python
#quantas linhas eu tenho para cada loja
# value counts
transacoes_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_loja)
```

```python
#agrupar os produtos e some as colunas de valor final
# group by
#coloca no groupby o que eu quero agrupar
#depois tenho que dizer o que quero fazer com as outras colunas numéricas
faturamento_produto = vendas_df.groupby('Produto').sum()
display(faturamento_produto)
```

```python
#para filtrar só as colunas que eu quero
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```
